```python
def calculate_jaccard_similarity(str1, str2):
    # Convert strings into sets of words
    set1 = set(str1.lower().split())
    set2 = set(str2.lower().split())

    # Calculate intersection and union
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))

    # Avoid division by zero
    if union == 0:
        return 0

    # Jaccard similarity formula
    similarity = intersection / union

    return similarity
```

```python
str1 = input("Enter the first string: ")
str2 = input("Enter the second string: ")
```

```python
similarity = calculate_jaccard_similarity(str1, str2)

print("Jaccard Similarity:", similarity)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Download required datasets (run once)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
text1 = input("Enter first text: ")
text2 = input("Enter second text: ")
```

```python
def text_similarity(text1, text2):

    # Convert to lowercase
    text1 = text1.lower()
    text2 = text2.lower()

    # Tokenization
    tokens1 = word_tokenize(text1)
    tokens2 = word_tokenize(text2)

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens1 = [lemmatizer.lemmatize(token) for token in tokens1]
    tokens2 = [lemmatizer.lemmatize(token) for token in tokens2]

    # Remove stopwords
    stop_words = stopwords.words('english')
    tokens1 = [token for token in tokens1 if token not in stop_words]
    tokens2 = [token for token in tokens2 if token not in stop_words]

    print("Processed Tokens 1:", tokens1)
    print("Processed Tokens 2:", tokens2)

    # Join tokens back into strings
    processed_text1 = " ".join(tokens1)
    processed_text2 = " ".join(tokens2)

    # TF-IDF Vectorization
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([processed_text1, processed_text2])

    # Cosine similarity
    similarity = cosine_similarity(vectors[0], vectors[1])

    return similarity[0][0]
```

```python
nltk.download('punkt_tab')
sim = text_similarity(text1, text2)

print("Cosine Similarity:", sim)
```
