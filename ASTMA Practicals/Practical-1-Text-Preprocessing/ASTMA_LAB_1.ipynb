```python
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
text = "Hello! This is a sample text with numbers 123 and symbols #@$."
print("Original Text:", text)
```

```python
tokens = word_tokenize(text)
print("Tokens:", tokens)
```

```python
tokens = [word.lower() for word in tokens]
print("Lowercase Tokens:", tokens)
```

```python
tokens = [word for word in tokens if word.isalpha()]
print("Alphabetic Tokens:", tokens)
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

print("Filtered Tokens:", filtered_tokens)
```

```python
def is_valid_script(word):
    return re.match("^[A-Za-z]+$", word) is not None
```

```python
valid_tokens = [word for word in filtered_tokens if is_valid_script(word)]

print("Valid Script Tokens:", valid_tokens)
```
