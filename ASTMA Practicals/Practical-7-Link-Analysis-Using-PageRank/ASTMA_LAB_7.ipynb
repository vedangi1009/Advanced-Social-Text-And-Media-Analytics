```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create directed graph
G = nx.DiGraph()

# Add edges (web links)
G.add_edges_from([
    ('A', 'B'),
    ('A', 'C'),
    ('B', 'C'),
    ('C', 'A'),
    ('D', 'C')
])

# Print nodes and edges
print("Nodes:", G.nodes())
print("Edges:", G.edges())
```

```python
plt.figure(figsize=(6,6))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=2000, font_size=12)
plt.title("Web Link Structure")
plt.show()
```

```python
pagerank = nx.pagerank(G, alpha=0.85)

print("PageRank Values:")
for page, rank in pagerank.items():
    print(page, ":", round(rank, 4))
```

```python
sorted_pr = sorted(pagerank.items(), key=lambda x: x[1], reverse=True)

print("\nFinal Page Ranking:")
for page, rank in sorted_pr:
    print(page, ":", round(rank, 4))
```
