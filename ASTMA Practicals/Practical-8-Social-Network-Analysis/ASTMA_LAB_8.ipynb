```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    'Source': ['A', 'A', 'B', 'C', 'D'],
    'Target': ['B', 'C', 'C', 'D', 'A']
}

df = pd.DataFrame(data)
df
```

```python
G = nx.from_pandas_edgelist(df, 'Source', 'Target', create_using=nx.DiGraph())

print("Nodes:", G.nodes())
print("Edges:", G.edges())
```

```python
plt.figure(figsize=(6,6))

pos = nx.spring_layout(G)  # better layout

nx.draw(G, pos, with_labels=True, node_color='lightgreen',
        node_size=2000, font_size=12, font_weight='bold', arrows=True)

plt.title("Social Network Graph")
plt.show()
```

```python
degree = nx.degree_centrality(G)
betweenness = nx.betweenness_centrality(G)

print("Degree Centrality:")
print(degree)

print("\nBetweenness Centrality:")
print(betweenness)
```

```python
# Highlight nodes based on degree centrality
node_size = [v * 3000 for v in degree.values()]

plt.figure(figsize=(6,6))
pos = nx.spring_layout(G)

nx.draw(G, pos,
        with_labels=True,
        node_size=node_size,
        node_color='skyblue',
        font_weight='bold',
        arrows=True)

plt.title("Network with Centrality Highlighted")
plt.show()
```
