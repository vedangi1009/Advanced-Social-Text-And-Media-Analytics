```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')        # WordNet
nltk.download('omw-1.4')        # Open Multilingual WordNet (OMW)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
text = "The striped bats are hanging on their feet and ate best fishes while running"
```

```python
nltk.download('punkt_tab')
tokens = word_tokenize(text.lower())
print("Tokens:", tokens)
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word not in stop_words]
print("After Stopword Removal:", filtered_tokens)
```

```python
ps = PorterStemmer()

stemmed_words = [ps.stem(word) for word in filtered_tokens]
print("After Stemming:", stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("After Lemmatization:", lemmatized_words)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
from nltk.corpus import wordnet

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

pos_tags = nltk.pos_tag(filtered_tokens)

lemmatized_pos = [
    lemmatizer.lemmatize(word, get_wordnet_pos(tag))
    for word, tag in pos_tags
]

print("Lemmatization with POS:", lemmatized_pos)
```
